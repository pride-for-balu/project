```python
import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
import pandas as pd
df = pd.read_csv("C://Users//BALU/ToyotaCorolla.csv", encoding='latin1')
```

```python
df.describe()
```

```python
df1 = df[["Price","Age_08_04","KM","HP","cc","Doors","Gears","Quarterly_Tax","Weight"]]
```

```python
df1
```

```python
X = df1.drop("Price", axis=1)
Y = df1.Price
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.30)
```

```python
# Apply multiple Linear Regression Model
reg = LinearRegression()
reg.fit(X_train,Y_train)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
# Generate Prediction on test set
reg_y_pred = reg.predict(X_test)
```

```python
# Calculating Mean Squared Error MSE
mean_squared_error = np.mean((reg_y_pred - Y_test)**2)
print("Mean Squared Error on test set : ", mean_squared_error)
```

```python
# putting together the coefficient and their corresponding variables names
reg_coefficient = pd.DataFrame()
reg_coefficient['columns'] = X_train.columns
reg_coefficient['coefficient Estimate'] = pd.Series(reg.coef_)
print(reg_coefficient)
```

```python
# import ridge regression from sklearn library
from sklearn.linear_model import Ridge

# Train model 
ridge = Ridge(alpha=2)
ridge.fit(X_train, Y_train)
y_pred = ridge.predict(X_test)

# calculate mean squared error

mean_squared_error_ridge = np.mean((y_pred - Y_test)**2)
print(mean_squared_error_ridge)

# putting together the coefficient and their corresponding variables names
ridge_reg_coefficient = pd.DataFrame()
ridge_reg_coefficient['columns'] = X_train.columns
ridge_reg_coefficient['coefficient Estimate'] = pd.Series(ridge.coef_)
print(ridge_reg_coefficient)
```

```python
y_pred_tr = ridge.predict(X_train)
y_pred_te = ridge.predict(X_test)
```

```python
from sklearn import metrics
print("RMSE FOR TRAIN DATA : ", np.sqrt(metrics.mean_squared_error(Y_train, y_pred_tr)))
print("R2 for train data", metrics.r2_score(Y_train, y_pred_tr))

print("RMSE FOR TEST DATA : " , np.sqrt(metrics.mean_squared_error(Y_test, y_pred_te)))
print("R2 for test data", metrics.r2_score(Y_test, y_pred_te))
```

```python
# model 
from sklearn.linear_model import ElasticNet

# Train the model
e_net = ElasticNet(alpha=.01, l1_ratio=.5)
e_net.fit(X_train, Y_train)

# calculate the prediction and mean square error 
y_pred_elastic = e_net.predict(X_test)
Mean_squared_error = np.mean((y_pred_elastic - Y_test)**2)
print("mean squared error on test set", Mean_squared_error)

# putting together the coefficient and their corresponding variables names
elastic_coefficient = pd.DataFrame()
elastic_coefficient['columns'] = X_train.columns
elastic_coefficient['coefficient Estimate'] = pd.Series(e_net.coef_)
print(elastic_coefficient)
```
